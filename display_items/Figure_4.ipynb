### *Localizing the effects of physiochemical features in Treg-Tconv differentiation*

```r
paste(path_to_results, "TCRfeat_effectsizes/physiochemical_features/seay2016/ridge_regularized", sep="/")
```

See *rrLR_allregions.R* for our ridge-regularized modeling of physiochemical feature effects.

```r
setwd(paste(path_to_results, "TCRfeat_effectsizes/physicochemical_features/seay2016/ridge_regularized", sep="/"))
files = list.files()
files = files[!(grepl("png", files))]
all = data.frame(matrix(nrow=0, ncol=4))
colnames(all) = c("length", "imgt_pos", "feature", "estimate")
for (i in 1:length(files)){
  res = readRDS(files[i])
  res$length = strsplit(files[i], "_")[[1]][1]
  res$imgt_pos = sapply(res$term, function(x) strsplit(x, "_")[[1]][1])
  res$feature = sapply(res$term, function(x) strsplit(x, "_")[[1]][2])
  all = rbind(all, res[,colnames(all)])
}
all = all[all$feature %in% c("volume", "pI", "hydrophob"),]
all$feature = as.character(all$feature)
all$feature[all$feature=="hydrophob"] = "hydrophobicity"
all$feature[all$feature=="pI"] = "isoelectric point"
all$imgt_pos = as.character(all$imgt_pos)
all$imgt_pos = factor(all$imgt_pos, levels=c("p27", "p28", "p29", "p37", "p38", "p56", "p57", "p58", "p63", "p64", "p65", "p104", "p105", "p106", "p107", "p108", "p109", "p110", "p111", "p111.1", "p112.1", "p112", "p113", "p114", "p115", "p116", "p117", "p118"))
```

```r
gr = all %>% group_by(imgt_pos, feature) %>% dplyr::summarise(coef = mean(estimate))
g = ggplot(gr, aes(imgt_pos, exp(coef), color=feature))
g = g + geom_point() + geom_line(aes(group=feature)) + theme_classic() + geom_hline(yintercept=1)
g = g + theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust=1)) + scale_color_manual(values=c("#eead0c", "#ed6a51", "#02868a"))
g + xlab("TCR position") + ylab("coefficient for Treg prediction")
```

### *Visualizing average CDR3bmr hydrophobicity in Tregs vs Tconvs*

```r
data = readRDS(paste(path_to_data, "seay2016/seay_training.rds", sep="/"))
```

```r
aa_maps = list()
rf = readRDS(paste(path_to_data, "aminoacid_physiochemical_feats.rds", sep="/"))
for (i in 2:ncol(rf)){
  aa_maps[[i-1]] = create_hashmap(as.character(rf$AA), as.numeric(as.character(rf[,i])))
}
data$mid_hydro = sapply(data$midseq, function(x) get_feat_score(x, aa_maps[[2]]))
data$smid_hydro = scale(data$mid_hydro)[,1]
```

```r
library(ggridges)
g = ggplot()
g = g + geom_vline(xintercept = mean(data$smid_hydro[data$cell_state=="Tconv"]), color="turquoise3")
g = g + geom_vline(xintercept = mean(data$smid_hydro[data$cell_state=="Treg"]), color="darkmagenta")
g = g + geom_density_ridges(aes(x=data$smid_hydro, y=data$cell_state, color=data$cell_state, fill=data$cell_state), bandwidth=0.5, alpha=0.4, show.legend = FALSE) + scale_color_manual(values=c("turquoise3", "darkmagenta"))
g = g + scale_fill_manual(values=c("turquoise3", "darkmagenta")) + theme_bw(base_size=12)
g + xlim(c(-4,4)) + xlab("CDR3bmr hydrophobicity") + ylab("") + coord_flip() 
```

### *TRBV gene effects correspond to p37 electrostatic charge*

```r
cdref = readRDS(paste(path_to_data, "CDR1bCDR2b_reference.rds", sep="/"))
cdref$vgene = gsub("TR", "TCR", cdref$vgene)
print(load(paste(path_to_results, "TCRfeat_effectsizes/Vregion/seay2016/seaytrain_vregion.RData", sep="/")))
#print(load("../final_training_betas_rmmx/seay70_all_VGvg_aa_res.RData"))
rare = names(tvg)[tvg/sum(tvg) < 0.005 ]
stats = stats[!(stats$term %in% paste("vgene", rare, sep="")),]
toplot = stats[grepl("vgene", stats$term),]
toplot$OR = exp(toplot$estimate)
toplot$minOR = exp(toplot$estimate + qnorm(0.025)*toplot$std.error)
toplot$maxOR = exp(toplot$estimate + qnorm(0.975)*toplot$std.error)
toplot$vgene = sapply(toplot$term, function(x) gsub("vgene", "", x))
toplot = left_join(toplot, cdref[,c("vgene", "p37", "p37_pi_feat")])
toplot$aagroup = "neutral"
toplot$aagroup[toplot$p37 %in% c("D", "E")] = "neg"
toplot$aagroup[toplot$p37 %in% c("K", "R")] = "pos"
toplot$aagroup[toplot$p37 %in% c("L")] = "hydro"

g = ggplot(toplot, aes(p37_pi_feat, OR, label=p37, color=aagroup))
g = g + geom_hline(yintercept=1, linetype="dashed") + geom_errorbar(aes(ymin=minOR, ymax=maxOR), color="gray") + geom_text(fontface="bold", show.legend=FALSE, size=7) + theme_classic(base_size=12) + scale_color_manual(values=c("darkgoldenrod2", "navyblue", "black", "coral2"))
g + xlab("p37 isoelectric point") + ylab("TRBV gene odds ratio")
```

### *Visualization of amino acid effects along the CDR3bmr*

```r
##upload this
npc = readRDS(paste(path_to_results, "TCRfeat_effectsizes/midregion_positionalAA/seay2016/noperccovs.rds", sep="/"))
sub = npc[npc$length=="15",]
sub$aa = as.character(sub$aa)
pos = c("108", "109", "110", "111", "112")

mat = matrix(nrow=length(aminos()), ncol=length(pos))
colnames(mat) = pos
rownames(mat) = aminos()
for (i in 1:nrow(mat)){
  for (j in 1:ncol(mat)){
    if (nrow(sub[sub$imgtpos==pos[j] & sub$aa==aminos()[i],])>0){
      mat[i,j] = sub$beta[sub$imgtpos==pos[j] & sub$aa==aminos()[i]]
    } else {
      mat[i,j] = 0
    }
  }
}

ggseqlogo(mat, method='custom', seq_type='aa', col_scheme=amino_color_scheme())
```
