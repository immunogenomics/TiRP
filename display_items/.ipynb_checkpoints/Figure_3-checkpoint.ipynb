### *Comparing amino acid usage in Tregs and Tconvs*

In the training dataset, we have pre-computed the percent occupancy of each amino acid in the CDR3bmr. See the **add_percentAA()** function in *utils.R* for full doucmentation.

Key computation in **add_percentAA():**

`sapply(data$midseq, function(x) 100*str_count(x, aminos[i])/nchar(x))`

```r
head(data[,c("sequence", "midseq", colnames(data)[grepl("perc", colnames(data))])])
```

```r
cs = amino_color_scheme()
options(dplyr.summarise.inform=FALSE)
```

```r
aminos_3a <- c("L", "F", "W", "Y", "R", "D", "E", "Q")
plots = list()
for (i in 1:length(aminos_3a)){
    ind = which(colnames(data)==paste("perc_mid", aminos_3a[i], sep="_"))
    data$perc_mid = data[,ind]
    toplot = data %>% group_by(cell_state, donor) %>% dplyr::summarise(mean_perc = mean(perc_mid))
    g = ggplot(toplot, aes(cell_state, mean_perc, group=donor))
    plots[[i]] = g + geom_point(color=cs$col[cs$letter==aminos_3a[i]]) + geom_line(color=cs$col[cs$letter==aminos_3a[i]]) + xlab("") + ylab("") + theme_bw(base_size=15) + ggtitle(aminos_3a[i])
}
```

```r
options(repr.plot.width=10, repr.plot.height=4)
wrap_plots(plots, ncol=4)
```

### *Visualizing the gain in variance explained with each additional TCR feature*

We used forward selection to build a mixed effects logistic regression model for each region of the TCR (see *meLR_midregion.R*, *meLR_Vregion.R*, and *meLR_Jregion.R.*).

To examine the gain in Treg-Tconv variance explained with each additional TCR feature, we calculated the psuedo R2 at each step of forward selection
(key computation:
`varex = 100 * (1-(an$logLik[2]/an$logLik[1]))`).

See *calculate_incremental_varex.R* for full documentation.

```r
dat = read.csv(paste(path_to_results, "variance_explained/incremental_varex_seaydisc.csv", sep="/"))
```

```r
head(dat)
```

```r
total_varex = max(dat$varex[dat$region=="V"]) + max(dat$varex[dat$region=="mid"]) + max(dat$varex[dat$region=="J"])
dat$ve_norm = 100 * (dat$varex/total_varex)
dat$type = "alternative"
dat$type[dat$series %in% c("vg", "mot_jgs", "AA")] = "primary"
```

```r
v = plot_varex(dat, "V")
m = plot_varex(dat, "mid")
j = plot_varex(dat, "J")
```

```r
v + m + j
```

### *Visualizing contributions by each TCR feature in the final model*

```r
all = readRDS(paste(path_to_results, "variance_explained/incremental_varex_bylength_seaydisc.rds", sep="/"))
```

```r
allg = all %>% group_by(length) %>% dplyr::summarise(total_varex = sum(varex))
all = left_join(all, allg, by="length")
all$ve_norm = 100 * (all$varex/all$total_varex)
g = ggplot(all, aes(length, ve_norm, fill=feat))
g = g + geom_bar(stat="identity", position="stack", show.legend = FALSE) + coord_flip() + theme_classic(base_size=12) + scale_fill_manual(values=c("#cb7392", "#ffc0cb", "#6959cd", "#cd96cd", "#8b636c", "#87ceff", "#00688b", "#5e4806", "#ffec8b", "#e7862b", "#9bcd9b", "#628d55", "#19361c"))
g + xlab("percent of explained variance")
```

```r
all$region = "mid"
all$region[all$feat %in% c("vgene", "p107", "VGS")] = "V"
all$region[all$feat %in% c("Jmotif", "p113")] = "J"
all$region_size = 11
all$region_size[all$region=="J"] = 6
all$region_size[all$region=="mid" & all$length==12] = 2
all$region_size[all$region=="mid" & all$length==13] = 3
all$region_size[all$region=="mid" & all$length==14] = 4
all$region_size[all$region=="mid" & all$length==15] = 5
all$region_size[all$region=="mid" & all$length==16] = 6
all$region_size[all$region=="mid" & all$length==17] = 7
all$region_prop = all$region_size/(as.numeric(as.character(all$length))+15)
```

```r
byregion = all %>% group_by(region, region_prop, length) %>% dplyr::summarise(varex = sum(varex))
g = ggplot(byregion, aes(region_prop, varex, color=region, label=length))
g = g + geom_smooth(method="lm", show.legend = FALSE) + geom_point(size=4, color="black") + geom_point(size=3.5) + geom_text(color="black", size=2.5) + theme_bw(base_size=15)  + scale_color_manual(values=c("#cb7392", "#e7862b", "#628d55"))
g + xlab("proportion of TCR") + ylab("variance explained")
```
