#### Labeling cell types based on Symphony mappings

Symphony maps query cells to coordinates in the reference's harmonized prinicpal component space. To infer cell types from these mappings, we find the dominant cell type of the five nearest neighbor reference cells (in harmonized PC space) for each query cell. K-nearest-neighbors is implemented in R by the package ```class```.

````r
scored$predicted_cluster <- knn(t(ref$Z_corr), t(Z), cl=clusters, k=5)
scored$predicted_cellstate <- "Tconv"
scored$predicted_cellstate[scored$predicted_cluster %in% c("6")] <- "Treg"

table(scored$predicted_cellstate)

clones <- scored %>% group_by(patient, dataset, ident, TiRP, mode_TRA) %>%
       dplyr::summarise(nTreg = n_distinct(barcode[predicted_cellstate=="Treg"]), 
                     nTconv = n_distinct(barcode[predicted_cellstate=="Tconv"]))
````

**Note:** output from ```class::knn()``` will slightly vary due to stochastic tie-breaking when there is no single dominant cell type among a query cell's nearest neighbors. 

````r
pts = as.matrix(harm_PCs[,2:ncol(harm_PCs)])
nbs = mcNNindex(target=pts, query=pts, k=100)
scored$TiRP_smoothed = sapply(1:nrow(scored), function(x) mean(scored$TiRP[nbs[x,]]))
scored$diff = sapply(scored$TiRP_smoothed, function(x) x-mean(scored$TiRP_smoothed))
scored = scored[order(abs(scored$diff)),]
````

````r
g = ggplot() + geom_point_rast(aes(x=rmd$UMAP_1, y=rmd$UMAP_2), color="black", size=0.0001)
g = g + geom_point_rast(aes(x=scored$UMAP_1, y=scored$UMAP_2, color=scored$TiRP_smoothed), size=0.2) + theme_classic() + scale_color_gradient2(low="turquoise3", midpoint=mean(scored$TiRP_smoothed), mid="gray95", high="darkmagenta")
g + xlab("UMAP 1") + ylab("UMAP 2") + labs(color="neighborhood TiRP score")
````

````r
clones$clonetype <- "mixed"
clones$clonetype[clones$nTreg==0] <- "Tconv"
clones$clonetype[clones$nTconv==0] <- "Treg"
````

````r
scored = left_join(scored, clones[,c("ident", "clonetype")])
````

````r
scored$clonetype = factor(scored$clonetype, levels=c("Tconv", "mixed", "Treg")) ##ordering clonetypes by frequency for optimal visualization
scored = scored[order(scored$clonetype),]
g = ggplot() + geom_point_rast(aes(x=rmd$UMAP_1, y=rmd$UMAP_2), color="black", size=0.0001)
g = g + geom_point_rast(aes(x=scored$UMAP_1, y=scored$UMAP_2, color=scored$clonetype), size=0.2) + theme_classic() + scale_color_manual(values=c("turquoise3", "darkgoldenrod2", "darkmagenta"))
g + xlab("UMAP 1") + ylab("UMAP 2") + labs(color="neighborhood TiRP score")
````

**Test 1**: Do mixed clones harbor higher TiRP scores than clones that maintain the Tconv phenotype through expansion?

````r
tc = clones[clones$clonetype!="Treg",]
tc$clonetype = factor(tc$clonetype, levels=c("Tconv", "mixed"))
fit = glmer(clonetype ~ TiRP + (1|patient), data=tc, family="binomial")
tidy(fit)
````

**Yes,** in these data we see significant evidence that mixed clones harbor higher TiRP scores than clones that maintain the Tconv phenotype through expansion.

**Test 2**: Do mixed clones harbor lower TiRP scores than those that maintain the Treg phenotype through expansion?

````r
tr = clones[clones$clonetype!="Tconv",]
tr$clonetype = factor(tr$clonetype, levels=c("mixed", "Treg"))
fit = glmer(clonetype ~ TiRP + (1|patient), data=tr, family="binomial")
tidy(fit)
````

While TiRP scores are higher in Treg expanded clones than in mixed clones, this difference is not statistically significant.

Note: the *P* values given above do not precisely match those given in **Figure 6g** because, as discussed above, ties are broken stochastically for ```class::knn()```, which we use to assign cell state labels from Symphony mappings. 

The general results, however, including the *P* value order of magnitude, are robust to random seed value.

````r
clones$size = clones$nTreg + clones$nTconv
clones$rat = (clones$nTreg + 1)/(clones$nTconv + 1)
````

````r
g = ggplot(clones, aes(TiRP, rat, color=log(rat), size=size))
g = g + geom_point() + theme_bw(base_size=15) + geom_smooth(method="lm", color="gray") + scale_y_continuous(limits=c(0.01, 11), trans='log', breaks=c(1/10, 1/25, 1/10, 1/5, 1/2, 1, 2/1, 5/1, 10/1))
g + scale_color_gradient2(low="turquoise3", high="darkmagenta", mid="darkgoldenrod2", midpoint=log(0.5)) + ylab("Treg/Tconv ratio")
````

**Test 3:** does TiRP help to explain the ratio of Tregs to Tconvs in expanded clones?

````r
fit = glmer(log(rat) ~ TiRP + (1|patient), data=clones)
fit0 = glmer(log(rat) ~ (1|patient), data=clones)
tidy(fit)
anova(fit, fit0)
````

**Yes**, in these data, one unit increase in TiRP (equal to one TiRP standard deviation in held-out bulk sequencing data) corresponds to a 6% rise in the ratio of Tregs to Tconvs in an expanded clone.

Once again, we see minor discrepancies from the exact values given in **Figure 6h** due to stochastic tie-breaking in ```class::knn()```.

The general results, however, including the P value order of magnitude, are robust to random seed value.
