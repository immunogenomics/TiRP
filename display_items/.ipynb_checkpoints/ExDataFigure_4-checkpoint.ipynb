```r
calc_VJ_selectionrate <- function(dat, gene="v_gene"){
  dat = dat[!(duplicated(dat$amino_acid)),]
  prod = dat[dat$frame_type=="In",]
  nonprod = dat[dat$frame_type!="In",]
  tC = nrow(prod) + 1
  tD = nrow(nonprod) + 1
  dg = dat %>% group_by(!!! rlang::syms(gene)) %>% dplyr::summarise(SR = log(((length(frame_type[frame_type=="In"])+1)/(length(frame_type[frame_type!="In"])+1))/(tC/tD)))
  return(dg)
}
```

```r
library(dplyr)
library(rlang)
library(ggplot2)
```

```r
print(load("/panfs/pan01/hpc/groups/srlab2/klagattuta/seay/seay_data.rds"))
```

```r
all = all[all$cell_state %in% c("Tconv", "Treg", "Tconv-a", "Tconv-b"),]
```

```r
head(all)
```

```r
all$amino_acid = as.character(all$sequence)
```

```r
all$v_gene = all$vgene
all$j_gene = all$jgene
```

```r
drs = unique(as.character(all$donor))
```

```r
vgsr = NULL
jgsr = NULL
```

```r
for (i in 1:length(drs)){
    dat = all[all$donor==drs[i],]
    v = calc_VJ_selectionrate(dat, "v_gene")
    j = calc_VJ_selectionrate(dat, "j_gene")
    vgsr = rbind(vgsr, cbind(donor=rep(drs[i], nrow(v)), v))
    jgsr = rbind(jgsr, cbind(donor=rep(drs[i], nrow(j)), j))
}
```

```r
head(vgsr)
```

```r
df = data.frame(donor = c(vgsr$donor, jgsr$donor), gene = c(vgsr$v_gene, jgsr$j_gene), SR = c(vgsr$SR, jgsr$SR), type=c(rep("V", nrow(vgsr)), rep("J", nrow(jgsr))))
```

```r
df$gene = as.character(df$gene)
df$type = as.character(df$type)
```

```r
df$gene[df$gene=="unresolved"] = paste("unresolved", df$type[df$gene=="unresolved"], sep="")
```

```r
df$type = factor(df$type, levels=c("V", "J"))
df = df[order(df$type, df$gene),]
df$gene = factor(df$gene, levels=unique(df$gene))
```

```r
options(repr.plot.width=15, repr.plot.height=5)
g = ggplot(df, aes(gene, SR, color=type))
g + geom_jitter(alpha=0.8, width=0.2, size=2, show.legend=FALSE) + theme_bw() + theme(axis.text.x = element_text(angle=45, vjust=1, hjust=1))
```
