```r
data = readRDS("../data/seay2016/seay_discovery.rds")
```

```r
plot_maxcorrs <- function(vars, cormat){
    maxs = data.frame(matrix(nrow=length(vars), ncol=length(vars)))
    for (i in 1:length(vars)){
        delimi = "is"
        if (vars[i]=="perc_mid") { delimi = "" }
        for (j in 1:length(vars)){
            delimj = "is"
            if (vars[j]=="perc_mid") { delimj = "" }
            if (i==j) { 
                maxs[i,j] = 1
            } else {
                mini = cormat[grepl(paste(vars[j], delimj, sep="_"), rownames(cormat)), grepl(paste(vars[i], delimi, sep="_"), colnames(cormat))]
                inds = which(abs(mini) == max(abs(mini)), arr.ind = TRUE)
                maxs[i,j] = mini[inds[1], inds[2]]
            }
        }
    }
    vars = gsub("perc_mid", "CDR3bmr %AAs", vars)
    vars = gsub("jgene", "TRBJ gene", vars)
    vars = gsub("vgene", "TRBV gene", vars)
    colnames(maxs) = vars
    rownames(maxs) = vars
    return(corrplot2(as.matrix(maxs), method="color", col=ocean.curl(100), addCoef.col = "white", addCoef.font = 1, type="lower", tl.col="black", tl.srt=45))           
}
```

```r
vars=c(colnames(data)[grepl("perc_mid", colnames(data))], "length", "p108", "p109", "p110", "p111", "p111.1", "p112.1", "p112")
cormat = get_cormat(vars, region="midregion", dataset="seay_disc")
plot_maxcorrs(c("perc_mid", vars[!(grepl("perc_mid", vars))]), cormat)
```

```r
vars = c("p113", "p114", "p115", "p116", "p117", "Jmotif", "jgene")
cormat = get_cormat(vars, region="Jregion", dataset="seay_disc")
plot_maxcorrs(vars, cormat)
```

```r
vars = c("p27", "p28", "p29", "p37", "p38", "p56", "p57", "p58", "p63", "p64", "p65", "vgene", "Vmotif", "p105", "p106", "p107")
cormat = get_cormat(vars, region="Vregion", dataset="seay_disc")
plot_maxcorrs(vars, cormat)
```
